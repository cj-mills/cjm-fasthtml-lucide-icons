```python
#| export
from dataclasses import dataclass, field
from typing import List, Dict, Optional, Union
```

## SVG Element Types

Dataclasses representing the different SVG element types used in Lucide icons.

```python
#| export
@dataclass
class PathElement:
    """SVG path element data."""
    d: str  # path data string
```

```python
#| export
@dataclass
class CircleElement:
    """SVG circle element data."""
    cx: str  # center x coordinate
    cy: str  # center y coordinate
    r: str   # radius
```

```python
#| export
@dataclass
class RectElement:
    """SVG rect element data."""
    x: str              # x position
    y: str              # y position
    width: str          # width
    height: str         # height
    rx: Optional[str] = None  # corner radius (optional)
```

```python
#| export
@dataclass
class LineElement:
    """SVG line element data."""
    x1: str  # start x
    y1: str  # start y
    x2: str  # end x
    y2: str  # end y
```

```python
#| export
@dataclass
class EllipseElement:
    """SVG ellipse element data."""
    cx: str  # center x coordinate
    cy: str  # center y coordinate
    rx: str  # x radius
    ry: str  # y radius
```

```python
#| export
@dataclass
class PolylineElement:
    """SVG polyline element data."""
    points: str  # space-separated coordinate pairs
```

```python
#| export
@dataclass
class PolygonElement:
    """SVG polygon element data."""
    points: str  # space-separated coordinate pairs
```

```python
#| export
# Union type for all SVG elements
SvgElement = Union[
    PathElement,
    CircleElement,
    RectElement,
    LineElement,
    EllipseElement,
    PolylineElement,
    PolygonElement
]
```

## Icon Data Container

```python
#| export
@dataclass
class IconData:
    """Container for a Lucide icon's SVG elements and metadata."""
    elements: List[SvgElement]          # SVG child elements
    categories: List[str] = field(default_factory=list)  # category assignments
    tags: List[str] = field(default_factory=list)        # searchable tags
```
